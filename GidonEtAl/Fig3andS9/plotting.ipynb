```python
import pickle
import matplotlib.pylab as plt
import numpy as np
```

```python
def load_vec(name):
    with open(f'{name}.npy', 'rb') as f:
        vec = np.load(f)
    return vec
```

```python
timevec = load_vec('timevec')
```

```python
XY = load_vec('XY')
Y = load_vec('Y')
X = load_vec('X')
```

```python
plt.style.use('seaborn-v0_8-talk')

fig, ax = plt.subplots()

#x, y, xy
ax.plot(timevec,XY, linewidth = 1.2, alpha = .85, label = 'Both Pathways Active')
ax.plot(timevec,X, linewidth = 1.2, alpha = .85, label = 'X Pathway Active')
ax.plot(timevec,Y, linewidth = 1.2, alpha = .85, label = 'Y Pathway Active')


# plt.axvline(x=2000, color = 'red', linewidth = 1, linestyle = 'dashed')
ax.set_title('XOR Functionality')
ax.set_ylabel('Voltage (mV)')
ax.set_xlabel('Time (ms)')
ax.legend()
```

```python
no_phase = load_vec('no_phase')
quarter_phase = load_vec('quarter_phase')
half_phase = load_vec('half_phase')

plt.style.use('seaborn-v0_8-talk')

fig, (ax1,ax2,ax3) = plt.subplots(3,1,sharex=True)

#x, y, xy

ax1.plot(timevec,half_phase, linewidth = 1.1, alpha = 1, label = '1/2-Phase Shift', color = 'orange')
ax2.plot(timevec,quarter_phase, linewidth = 1.1, alpha = 1, label = '1/4-Phase Shift', color = 'blue')
ax3.plot(timevec,no_phase, linewidth = 1, alpha = 1, label = '0-Phase Shift', color = 'black')

ax1.axvline(x=2002.5, color = 'red', linewidth = 1, linestyle = 'dashed')
ax2.axvline(x=2001.25, color = 'red', linewidth = 1, linestyle = 'dashed')
ax3.axvline(x=2000, color = 'red', linewidth = 1, linestyle = 'dashed')

ax1.set_title('Phase Shifted Y Pathway Activation')
ax2.set_ylabel('Voltage (mV)')
ax3.set_xlabel('Time (ms)')

ax1.legend(loc="lower right")
ax2.legend(loc="lower right")
ax3.legend(loc="lower right")
```

```python
hz5 = load_vec('5hz')
hz10 = load_vec('10hz')
hz15 = load_vec('15hz')

plt.style.use('seaborn-v0_8-talk')

fig, ax = plt.subplots()

#x, y, xy
ax.plot(timevec,hz15, linewidth = 1.2, alpha = .85, label = '15 Hz')
ax.plot(timevec,hz10, linewidth = 1.2, alpha = .85, label = '10 Hz')
ax.plot(timevec,hz5, linewidth = 1.2, alpha = .85, label = '5 Hz')
ax.axvline(x=2000, color = 'red', linewidth = 1, linestyle = 'dashed')


# plt.axvline(x=2000, color = 'red', linewidth = 1, linestyle = 'dashed')
ax.set_title('Synapse Frequency Sensitivity')
ax.set_ylabel('Voltage (mV)')
ax.set_xlabel('Time (ms)')
ax.legend()
```
